# Linear-combinations

```python
#general libraries to include
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define vectors
u = np.array([1, 2])
v = np.array([3, 1])

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the original vectors
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label='u = [1, 2]')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='v = [3, 1]')

# Generate and plot linear combinations
for a in np.linspace(-2, 2, num=5):
    for b in np.linspace(-2, 2, num=5):
        w = a * u + b * v
        ax.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(color='gray', linestyle='--', linewidth=0.5)
ax.legend()
plt.title('Linear Combinations of Vectors u and v in 2D')
plt.show()
```

```python


# Define vectors
u = np.array([1, 0, 0])
v = np.array([0, 1, 0])
w = np.array([0, 0, 1])

# Create a figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the original vectors
ax.quiver(0, 0, 0, u[0], u[1], u[2], color='r', label='u = [1, 0, 0]')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='b', label='v = [0, 1, 0]')
ax.quiver(0, 0, 0, w[0], w[1], w[2], color='g', label='w = [0, 0, 1]')

# Generate and plot linear combinations
for a in np.linspace(-1, 1, num=5):
    for b in np.linspace(-1, 1, num=5):
        for c in np.linspace(-1, 1, num=5):
            comb = a * u + b * v + c * w
            ax.quiver(0, 0, 0, comb[0], comb[1], comb[2], color='gray', alpha=0.5)

ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_zlim([-1.5, 1.5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.title('Linear Combinations of Vectors in 3D')
plt.show()
```
