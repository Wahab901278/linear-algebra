# dependent-and-independent-vectors

first of all look at what are <b>dependent vectors</b>:
    <b>Dependent vectors</b> are those which can be made by some possible combinations of other vectors.
    e.g let's say there are three vectors u, v , w and then w is a dependent vector if w can be made by combination     of u and v.
    Mathematically, <b>w= au+bv</b>, where a and b are scalars.
        In general for 2-d plane with x and y components, if u and v lies on same line then they are linearly               dependent because they are not bringing any new thing to us.
    

 Now, what are <b>Independent vectors</b>:
    <b>InDependent vectors</b> are those which can be made by some possible combinations of other vectors.
        e.g let's say there are three vectors u, v , w and then w is a Independent vector if w can't be made by             combination of u and v.
    Mathematically, <b>w â‰  au+bv</b>, where a and b are scalars.
        In general for 2-d plane with x and y components, if u and v don't lie on same line then they are linearly         Independent because they are bringing any new thing to us and we'll see in code what does that mean.

    Linearly Independent Set of Vectors: A set of vectors is linearly independent if no vector in the set can be written as a linear combination of the others. In simpler terms, if you have a set of vectors, none of them are redundant; each vector adds new information or direction to the set. For example, the vectors {(1,0),(0,1)}{(1,0),(0,1)} in 2D space form a linearly independent set because neither vector can be written as a multiple of the other.

    Linearly Dependent Set of Vectors: Conversely, a set of vectors is linearly dependent if at least one vector in the set can be represented as a linear combination of the others. This means that there's redundancy in the set, and one or more vectors can be "created" by combining others. For instance, the vectors {(1,0),(2,0)}{(1,0),(2,0)} in 2D space are linearly dependent because the second vector is just twice the first vector.

```python
#general libraries to include
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define vectors
vec_dep_2d_1 = np.array([2, 4])
vec_dep_2d_2 = np.array([1, 2])

vec_ind_2d_1 = np.array([2, 4])
vec_ind_2d_2 = np.array([4, 1])

# Create a figure and axis
plt.figure(figsize=(10, 5))

# Dependent vectors in 2D
plt.subplot(1, 2, 1)
plt.quiver(0, 0, vec_dep_2d_1[0], vec_dep_2d_1[1], angles='xy', scale_units='xy', scale=1, color='r', label='vec_dep_2d_1')
plt.quiver(0, 0, vec_dep_2d_2[0], vec_dep_2d_2[1], angles='xy', scale_units='xy', scale=1, color='b', label='vec_dep_2d_2')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Dependent Vectors in 2D')
plt.legend()

# Independent vectors in 2D
plt.subplot(1, 2, 2)
plt.quiver(0, 0, vec_ind_2d_1[0], vec_ind_2d_1[1], angles='xy', scale_units='xy', scale=1, color='r', label='vec_ind_2d_1')
plt.quiver(0, 0, vec_ind_2d_2[0], vec_ind_2d_2[1], angles='xy', scale_units='xy', scale=1, color='b', label='vec_ind_2d_2')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Independent Vectors in 2D')
plt.legend()

plt.show()
```

```python
# Define vectors
vec_dep_3d_1 = np.array([1, 2, 3])
vec_dep_3d_2 = np.array([2, 4, 6])
vec_dep_3d_3 = np.array([3, 6, 9])

vec_ind_3d_1 = np.array([1, 0, 0])
vec_ind_3d_2 = np.array([0, 1, 0])
vec_ind_3d_3 = np.array([0, 0, 1])

# Create a figure
fig = plt.figure(figsize=(10, 10))

# Dependent vectors in 3D
ax1 = fig.add_subplot(211, projection='3d')
ax1.quiver(0, 0, 0, vec_dep_3d_1[0], vec_dep_3d_1[1], vec_dep_3d_1[2], color='r', label='vec_dep_3d_1')
ax1.quiver(0, 0, 0, vec_dep_3d_2[0], vec_dep_3d_2[1], vec_dep_3d_2[2], color='b', label='vec_dep_3d_2')
ax1.quiver(0, 0, 0, vec_dep_3d_3[0], vec_dep_3d_3[1], vec_dep_3d_3[2], color='g', label='vec_dep_3d_3')
ax1.set_xlim([-10, 10])
ax1.set_ylim([-10, 10])
ax1.set_zlim([-10, 10])
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title('Dependent Vectors in 3D')
ax1.legend()

# Independent vectors in 3D
ax2 = fig.add_subplot(212, projection='3d')
ax2.quiver(0, 0, 0, vec_ind_3d_1[0], vec_ind_3d_1[1], vec_ind_3d_1[2], color='r', label='vec_ind_3d_1')
ax2.quiver(0, 0, 0, vec_ind_3d_2[0], vec_ind_3d_2[1], vec_ind_3d_2[2], color='b', label='vec_ind_3d_2')
ax2.quiver(0, 0, 0, vec_ind_3d_3[0], vec_ind_3d_3[1], vec_ind_3d_3[2], color='g', label='vec_ind_3d_3')
ax2.set_xlim([-1.5, 1.5])
ax2.set_ylim([-1.5, 1.5])
ax2.set_zlim([-1.5, 1.5])
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_title('Independent Vectors in 3D')
ax2.legend()

plt.show()
```

## Dependent and independent sets of vectors

    Linearly Independent Set of Vectors: A set of vectors is linearly independent if no vector in the set can be written as a linear combination of the others. In simpler terms, if you have a set of vectors, none of them are redundant; each vector adds new information or direction to the set. For example, the vectors {(1,0),(0,1)}{(1,0),(0,1)} in 2D space form a linearly independent set because neither vector can be written as a multiple of the other.

    Linearly Dependent Set of Vectors: Conversely, a set of vectors is linearly dependent if at least one vector in the set can be represented as a linear combination of the others. This means that there's redundancy in the set, and one or more vectors can be "created" by combining others. For instance, the vectors {(1,0),(2,0)}{(1,0),(2,0)} in 2D space are linearly dependent because the second vector is just twice the first vector.

```python
def plot_vectors(vectors, title):
    fig, ax = plt.subplots()
    ax.set_title(title)
    ax.axvline(x=0, color='grey', lw=1)
    ax.axhline(y=0, color='grey', lw=1)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    for i, vec in enumerate(vectors):
        ax.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color='b', label=f'vec{i+1}')
    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.grid(True)
    ax.legend()
    plt.show()


```

```python
vectors_independent = np.array([[1, 0], [0, 1]])
plot_vectors(vectors_independent, "Linearly Independent Vectors")
```

```python
vectors_dependent = np.array([[1, 0], [2, 0]])
plot_vectors(vectors_dependent, "Linearly Dependent Vectors")
```
