# Linear combinations of dependent and independent vectors:-

## Dependent Vectors:-

uptill now , we have seen what are dependent and independent vectors, let's have a look what and how a linear combination of dependent vectors look like.
As, we know that in dependent vector , atleast one vector is linearly dependent on the other. It means that whether both of the vectors are equal or they lie on the same plane. So, the possible combination of them would also be on the same plane and <b>they aren't able to cover the whole space</b> for sure.

## Independent Vectors:-

now , let's have a look what and how a linear combination of Independent vectors look like.
As, we know that in Independent vector , all of the vectors are not linearly dependent on the other. It means they will not lie on the same plane.<b>So, the possible combination of them would be able to cover the whole space</b> for sure.

#### Note : When you are dealing with high number of vectors, it is good to denote them as a single dot which represents a single vector originating from the origin

```python
#general libraries to include
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define vectors
vec_dep_2d_1 = np.array([1, 2])
vec_dep_2d_2 = np.array([2, 4])

vec_ind_2d_1 = np.array([1, 2])
vec_ind_2d_2 = np.array([2, -1])

# Create grid for linear combination
grid_range = np.linspace(-5, 5, 10)
U, V = np.meshgrid(grid_range, grid_range)

# Linear combination of dependent vectors
lin_comb_dep_x = U * vec_dep_2d_1[0] + V * vec_dep_2d_2[0]
lin_comb_dep_y = U * vec_dep_2d_1[1] + V * vec_dep_2d_2[1]

# Linear combination of independent vectors
lin_comb_ind_x = U * vec_ind_2d_1[0] + V * vec_ind_2d_2[0]
lin_comb_ind_y = U * vec_ind_2d_1[1] + V * vec_ind_2d_2[1]

plt.figure(figsize=(10, 5))

# Plot dependent vectors and their linear combinations
plt.subplot(1, 2, 1)
plt.quiver(0, 0, vec_dep_2d_1[0], vec_dep_2d_1[1], angles='xy', scale_units='xy', scale=1, color='r', label='vec_dep_2d_1')
plt.quiver(0, 0, vec_dep_2d_2[0], vec_dep_2d_2[1], angles='xy', scale_units='xy', scale=1, color='b', label='vec_dep_2d_2')
plt.plot(lin_comb_dep_x, lin_comb_dep_y, 'k.')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Linear Combinations of Dependent Vectors in 2D')
plt.legend()

# Plot independent vectors and their linear combinations
plt.subplot(1, 2, 2)
plt.quiver(0, 0, vec_ind_2d_1[0], vec_ind_2d_1[1], angles='xy', scale_units='xy', scale=1, color='r', label='vec_ind_2d_1')
plt.quiver(0, 0, vec_ind_2d_2[0], vec_ind_2d_2[1], angles='xy', scale_units='xy', scale=1, color='b', label='vec_ind_2d_2')
plt.plot(lin_comb_ind_x, lin_comb_ind_y, 'k.')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Linear Combinations of Independent Vectors in 2D')
plt.legend()

plt.show()
```

```python
# Define linearly dependent vectors in 3D space
vec_dep_3d_1 = np.array([1, 2, 3])
vec_dep_3d_2 = np.array([2, 4, 6])  # This is 2 times vec_dep_3d_1

# Generate linear combinations of the dependent vectors
coefficients = np.linspace(-10, 10, 100)
lin_comb_dep = np.array([a * vec_dep_3d_1 + b * vec_dep_3d_2 for a in coefficients for b in coefficients])

# Extract x, y, z coordinates
lin_comb_dep_x = lin_comb_dep[:, 0]
lin_comb_dep_y = lin_comb_dep[:, 1]
lin_comb_dep_z = lin_comb_dep[:, 2]

# Plotting
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(111, projection='3d')

# Plot the vectors
ax1.quiver(0, 0, 0, vec_dep_3d_1[0], vec_dep_3d_1[1], vec_dep_3d_1[2], color='r', label='vec_dep_3d_1')
ax1.quiver(0, 0, 0, vec_dep_3d_2[0], vec_dep_3d_2[1], vec_dep_3d_2[2], color='b', label='vec_dep_3d_2')

# Plot the linear combinations
ax1.plot(lin_comb_dep_x, lin_comb_dep_y, lin_comb_dep_z, 'k.')

# Set plot limits
ax1.set_xlim([-30, 30])
ax1.set_ylim([-30, 30])
ax1.set_zlim([-30, 30])

# Labels
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.set_zlabel('Z axis')
plt.title('Linear Combinations of Dependent Vectors in 3D')
# Legend
ax1.legend()

# Show plot
plt.show()
```

### The vectors used below are very special because when we put these vectors into a matrix it would be non invertible which means that we can't get back "x" when we do Ax=b when b=[0 0 0 ]

```python
# Define the vectors
u = np.array([1, -1, 0])
v = np.array([0, 1, -1])
w = np.array([-1, 0, 1])

# Define points for the plane formed by u and v
point = np.array([0, 0, 0])
d1 = u
d2 = v

# Create a meshgrid for the plane
a = np.linspace(-1, 1, 10)
b = np.linspace(-1, 1, 10)
a, b = np.meshgrid(a, b)
x = point[0] + a * d1[0] + b * d2[0]
y = point[1] + a * d1[1] + b * d2[1]
z = point[2] + a * d1[2] + b * d2[2]

# Plotting
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')


# Plot the vectors
ax.quiver(0, 0, 0, u[0], u[1], u[2], color='r', label='u')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='b', label='v')
ax.quiver(0, 0, 0, w[0], w[1], w[2], color='g', label='w')

# Plot linear combination of u and v showing w
comb = -1 * u - 1 * v
ax.quiver(0, 0, 0, comb[0], comb[1], comb[2], color='k', linestyle='--', label='-1*u - 1*v')

# Labels and limits
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.legend()
plt.title('Linear Combinations of Dependent Vectors in 3D')
plt.show()
```

```python
# Define three linearly independent vectors
u = np.array([1, 0, 0])
v = np.array([0, 1, 0])
w = np.array([0, 0, 1])

# Define the range for linear combinations
a = np.linspace(-1, 1, 10)
b = np.linspace(-1, 1, 10)
c = np.linspace(-1, 1, 10)

# Create a meshgrid for the combinations
a, b, c = np.meshgrid(a, b, c)
x = a * u[0] + b * v[0] + c * w[0]
y = a * u[1] + b * v[1] + c * w[1]
z = a * u[2] + b * v[2] + c * w[2]

# Plotting
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors
ax.quiver(0, 0, 0, u[0], u[1], u[2], color='r', label='u')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='b', label='v')
ax.quiver(0, 0, 0, w[0], w[1], w[2], color='g', label='w')

# Plot the combinations
ax.scatter(x, y, z, color='k', alpha=0.1)

# Labels and limits
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.legend()
plt.title("linear combination of independent vectots in 3D")
plt.show()
```
