# Vector Space

What is a vector space...?

So, A vector space is a bunch of vectors and the rule is you are able to take their combinations + some rules which i believe aren't quite important though they are....

let's have a look on proper definition.

A vector space (or linear space) is a collection of vectors. A vector is an object that has both a magnitude (length) and a direction. The space must follow certain rules (axioms) for addition and scalar multiplication.

#### Axioms

   -> Vector Addition: Adding two vectors in the space results in another vector in the same space.
   
   -> Scalar Multiplication: Multiplying a vector by a scalar (number) results in another vector in the same space.
   
   -> Zero Vector: There is a special vector (zero vector) which, when added to any vector, does not change the vector.
   
   -> Additive Inverses: Every vector has an inverse such that when added together, they result in the zero vector.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

u = np.array([1, 1])
v = np.array([2, 3])
w = u + v
scalar = 2
v_scaled = scalar * v #notice here we are doing scalar multiplication as per our axioms.

plt.figure()
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label='u = [1, 1]')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='g', label='v = [2, 3]')
plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='b', label='u + v = [3, 4]')
plt.quiver(0, 0, v_scaled[0], v_scaled[1], angles='xy', scale_units='xy', scale=1, color='m', label='2 * v = [6, 4]')

plt.xlim(0, 7)
plt.ylim(0, 5)
plt.grid()
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Vector Space Visualization')
plt.show()
```

### Let's have a look where we'll deal with each axiom

#### 1. Closure under Addition

```python
v = np.array([2, 3])
u = np.array([1, 1])
resultant = v + u

plt.figure()
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='v = [2, 3]')
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='g', label='u = [1, 1]')
plt.quiver(0, 0, resultant[0], resultant[1], angles='xy', scale_units='xy', scale=1, color='b', label='v + u = [3, 4]')

plt.xlim(0, 4)
plt.ylim(0, 5)
plt.grid()
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Closure under Addition')
plt.show()
```

#### 2. Closure under Scalar Multiplication

```python
v = np.array([2, 3])
scalar = 2
scaled_vector = scalar * v

plt.figure()
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='v = [2, 3]')
plt.quiver(0, 0, scaled_vector[0], scaled_vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='2 * v = [4, 6]')

plt.xlim(0, 5)
plt.ylim(0, 7)
plt.grid()
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Closure under Scalar Multiplication')
plt.show()
```

#### 3. Commutativity of Vector Addition

```python
v = np.array([2, 3])
u = np.array([1, 1])
result1 = v + u
result2 = u + v

plt.figure()

plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='v = [2, 3]')
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='g', label='u = [1, 1]')
plt.quiver(0, 0, result1[0], result1[1], angles='xy', scale_units='xy', scale=1, color='b', label='v + u = [3, 4]')
plt.quiver(0, 0, result2[0], result2[1], angles='xy', scale_units='xy', scale=1,linestyle="----", color='black', label='u + v = [3, 4]')

plt.xlim(0, 4)
plt.ylim(0, 5)
plt.grid()
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Commutativity of Vector Addition')
plt.show()
```

```python
u = np.array([1, 1])
v = np.array([2, 3])
w = np.array([3, 1])
result1 = (u + v) + w
result2 = u + (v + w)

plt.figure()
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label='u = [1, 1]')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='g', label='v = [2, 3]')
plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='b', label='w = [3, 1]')
plt.quiver(0, 0, result1[0], result1[1], angles='xy', scale_units='xy', scale=1, color='m', label='(u + v) + w = [6, 5]')
plt.quiver(0, 0, result2[0], result2[1], angles='xy', scale_units='xy', scale=1, color='c', label='u + (v + w) = [6, 5]')

plt.xlim(0, 7)
plt.ylim(0, 6)
plt.grid()
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Associativity of Vector Addition')
plt.show()
```

## Not a vector space

To illustrate an example that is not a vector space, we need to show a scenario where one or more of the axioms defining a vector space are violated. Let's consider an example where closure under vector addition is violated. This can occur if we restrict the vectors to a certain region and then show that adding two vectors can result in a vector outside that region.

```python

v = np.array([2, 1])
u = np.array([-1, 3])
resultant = v + u

plt.figure()

plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='v = [2, 1]')
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='g', label='u = [-1, 3]')
plt.quiver(0, 0, resultant[0], resultant[1], angles='xy', scale_units='xy', scale=1, color='b', label='v + u = [1, 4]')

plt.xlim(-2, 4)
plt.ylim(0, 5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(which='both')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Example of a Set Not Forming a Vector Space')
plt.show()
```
