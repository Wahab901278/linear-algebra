## Column space

Column space in general gives you the span of the columns i.e all linear combinations of the columns.
Suppose we have a matrix A then Column space of A is a subspace of R^4 and this space got subspaces (columns in it).
Column space is also called range or image of the matrix

#### But how does the whole column space looks like?
#### Is this a whole 4-dimensional space or it is a subspace inside of the whole 4-D?

The above question's answer can be obtained with the aid of equations..
### Does Ax=b have a solutions for every R.H.S b?
The Answer is No, We can't fill the whole space with this as we can't find every R.H.S b with the combination of A.
### But Which R.H.S b's allow me to solve this?
So, I can solve Ax=b exactly when the R.H.S B is a vector in the column space and if b is not a combination of the columns, then there will be no x.

let's see how a column span looks like

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

A = np.array([[1, 2],
              [3, 6]])
c1 = A[:, 0]
c2 = A[:, 1]

fig, ax = plt.subplots()
ax.quiver(0, 0, c1[0], c1[1], angles='xy', scale_units='xy', scale=1, color='g', label='Column vector $\\mathbf{c}_1$')
ax.quiver(0, 0, c2[0], c2[1], angles='xy', scale_units='xy', scale=1, color='orange', label='Column vector $\\mathbf{c}_2$')

x_col = np.linspace(-10, 10, 100)
y_col = (c1[1] / c1[0]) * x_col

ax.plot(x_col, y_col, 'g--', label='Column space')

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(color='gray', linestyle='--', linewidth=0.5)
ax.legend()

plt.title('Column Space of Matrix A')
plt.show()
```

### Null space (Kernel)

```python

from mpl_toolkits.mplot3d import Axes3D

# Define the vector that spans the null space
null_vector = np.array([1, 1, -1])

# Create a figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the null space vector
ax.quiver(0, 0, 0, null_vector[0], null_vector[1], null_vector[2], color='r', label='Null space vector')

# Draw the line that represents the null space
t = np.linspace(-10, 10, 100)
x = t * null_vector[0]
y = t * null_vector[1]
z = t * null_vector[2]

ax.plot(x, y, z, 'b--', label='Null space')

# Set plot limits
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])

# Add labels and legend
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$x_3$')
ax.legend()

plt.title('Null Space of Matrix A')
plt.show()
```

```python
A = np.array([[1, 2],
              [3, 6]])
c1 = A[:, 0]
c2 = A[:, 1]

# \this vector spans the null space
null_vector = np.array([-2, 1])

# Plot the column vectors
fig, ax = plt.subplots()
ax.quiver(0, 0, c1[0], c1[1], angles='xy', scale_units='xy', scale=1, color='g', label='Column vector $\\mathbf{c}_1$')
ax.quiver(0, 0, c2[0], c2[1], angles='xy', scale_units='xy', scale=1, color='orange', label='Column vector $\\mathbf{c}_2$')


x_col = np.linspace(-10, 10, 100)
y_col = (c1[1] / c1[0]) * x_col

ax.plot(x_col, y_col, 'g--', label='Column space')


ax.quiver(0, 0, null_vector[0], null_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector spanning null space')

# the line that represents the null space
x_null = np.linspace(-10, 10, 100)
y_null = (null_vector[1] / null_vector[0]) * x_null

ax.plot(x_null, y_null, 'b--', label='Null space')


ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)


ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
ax.legend()

plt.title('Column Space and Null Space of Matrix A')
plt.show()
```
